```python
def get_index(tstart, tdata):
    j = 1
    for i in range(0,len(tdata)):
        if tstart > tdata[i]:
            j=i
            
    return j

def get_tpeak(tstart, tdata, ldata):
    j = get_index(tstart, tdata)
    k = j
    Lpeak = max(ldata[j:len(ldata)])
    
    for i in range(j, len(ldata)):
        if ldata[i] == Lpeak:
            k = i
            
    return tdata[k]

def model(ti, M_Ni, T0): 
    time = ti/86400
    L = M_Ni * Msun * (3.24e10*np.exp(-time/t_Ni) + 7.29e9*np.exp(-time/t_Co))
    L_trap = L * (1 - np.exp(-np.power(T0/time, 2)))
    return L_trap

def objective(params, x, y_observed, y_uncertainty, model_function):
    """
    params is [a, b]
    (y_observed - y_theory)^2 / y_uncertainty^2
    """
    
    return ((y_observed - model_function(x,params[0],params[1]))**2 / y_uncertainty**2).sum()

def get_Mej(T0, Ekin):
    Mej = T0 * day * np.power(Ekin/(C*kappa_gamma), 0.5)
    return Mej/Msun

def get_t_thin(kappa, Mej, vej):
    t = np.power(kappa*Mej, 0.5)/vej 
    return t

def get_t_rise(kappa, Mej, Ekin): 
    t = np.power(kappa/(beta*c), 0.5)*np.power((6/5)*(Mej**3/Ekin),0.25)
    return t/day
```

```python
t_peaks = np.zeros(len(Lum))
L_peaks_true = np.zeros(len(Lum))
L_peaks_calc = np.zeros(len(Lum))
L_peaks_calc2 = np.zeros(len(Lum))
L_peaks_calc3 = np.zeros(len(Lum))

def get_L_peak_kk(tau_Ni, tau_Co, beta, M_Ni):
    x = (2*e_Ni*M_Ni)/(np.power(beta,2)*np.power(tau_Ni,2))
    L = x*(0.83*(1 - beta*tau_Ni)*np.exp(-beta*tau_Ni)+26.56*(1-(1+beta*tau_Co)*np.exp(-beta*tau_Co)))
    return L
```
